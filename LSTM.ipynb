```python
file =  open("F:/ETL/Dataset/RHTDM.txt",mode = "r")
Q = file.read()
```

```python
Q
```

```python
from re import sub
Q = sub("\n"," ",Q)
Q = sub('\?','',Q)
Q = sub('  ',' ',Q)
```

```python
Q
```

```python
Q = Q.upper().split(' ')
```

```python
Q
```

```python
inp = []
op = []
for i in range(3,len(Q),1):
    w1 = Q[i-3]
    w2 = Q[i-2]
    w3 = Q[i-1]
    w4 = Q[i]
    inp.append(w1 + " " + w2 + " " + w3)
    op.append(w4)
```

```python
# inp
# op
```

```python
import pandas as pd
Q1 = pd.DataFrame([inp,op]).T
Q1.columns = ["inp","op"]
```

```python
Q1.head()
```

```python
len(inp)
```

```python
from numpy import unique
unique_words = list(unique(Q))
len(unique_words)
```

```python
print(unique_words)
```

```python
from numpy import zeros
X = zeros((Q1.shape[0],len(unique_words),(Q1.shape[1]+1)))
```

```python
for line_no,line in enumerate(inp):
    for word_no,word in enumerate(line.split(" ")):
        X[line_no,unique_words.index(word),word_no] = 1
```

```python
X[8,47,2]
```

```python
Y = zeros((283,92))
for line_no,word in enumerate(list(Q1.op)):
    Y[line_no,unique_words.index(word)] = 1
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,LSTM
from keras.layers.core import Activation
```

```python
nn = Sequential()
nn.add(LSTM(100,input_shape=(92,3)))
nn.add(Dense(92))
nn.add(Dropout(0.2))
nn.add(Dense(92))
nn.add(Dropout(0.2))
nn.add(Dense(92))
nn.add(Activation("softmax"))
```

```python
nn.compile(loss="categorical_crossentropy",metrics="accuracy",optimizer="adam")
model = nn.fit(X,Y,validation_split=0.1,validation_batch_size=20,epochs=200)
```

```python
acc = model.history['accuracy']
val_acc = model.history['val_accuracy']
```

```python
import matplotlib.pyplot as plt
plt.plot(acc)
plt.plot(val_acc)
```
